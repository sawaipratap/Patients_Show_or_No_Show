```python
import pandas as pd
import ml_utils as mt
```

```python
data=pd.read_csv("Medical_Appoint_no_Shows.csv")
```

```python
date_diff_cols=['ScheduledDay','AppointmentDay']
date_components_cols=['ScheduledDay','AppointmentDay']
simple_num_cols=['Age', 'Scholarship', 'Hipertension',
       'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']
cat_to_dummies_cols=[ 'Neighbourhood','Gender']
```

```python
dp=mt.DataPipe(cat_to_dummies=cat_to_dummies_cols,
              date_diffs=date_diff_cols,
              date_components=date_components_cols,
              simple_num=simple_num_cols)
```

```python
dp.fit(data)
```

```python
nhbrhood_clubbed_cats=dp.data_pipe.transformer_list[0][1].steps[2][1].var_cat_dict['Neighbourhood'][0]
```

```python
# nhbrhood_clubbed_cats
```

```python
data_tf=dp.transform(data)
```

```python
data_summary_data=data.copy()
```

```python
data_summary_data['Neighbourhood']=data_summary_data['Neighbourhood'].replace(
                                                            dict.fromkeys(nhbrhood_clubbed_cats,'__other__')
                                                                            )
```

```python
date_ScheduledDay=pd.to_datetime(data_summary_data['ScheduledDay'])
data_summary_data['ScheduledDay_Month']=date_ScheduledDay.dt.month
data_summary_data['ScheduledDay_Day_Of_Month']=date_ScheduledDay.dt.day
data_summary_data['ScheduledDay_Day_Of_Week']=date_ScheduledDay.dt.dayofweek
```

```python
date_AppointmentDay=pd.to_datetime(data_summary_data['AppointmentDay'])
data_summary_data['AppointmentDay_Month']=date_AppointmentDay.dt.month
data_summary_data['AppointmentDay_Day_Of_Month']=date_AppointmentDay.dt.day
data_summary_data['AppointmentDay_Day_Of_Week']=date_AppointmentDay.dt.dayofweek
```

```python
for col in ['Scholarship', 'Hipertension','Diabetes', 'Alcoholism',  'SMS_received']:
    
    data_summary_data[col]=data_summary_data[col].replace({0:'No',1:'Yes'})
```

```python
data_summary_data
```

```python
weekday_mapping = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
for col in ['ScheduledDay_Day_Of_Week','AppointmentDay_Day_Of_Week']:
    data_summary_data[col]=data_summary_data[col].map(weekday_mapping)
```

```python
month_mapping = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
for col in ['ScheduledDay_Month','AppointmentDay_Month']:
    data_summary_data[col]=data_summary_data[col].map(month_mapping)
```
